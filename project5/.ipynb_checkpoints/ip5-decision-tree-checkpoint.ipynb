### Project Description
This project uses a heart disease factors dataset to predict if a patient is at risk of heart disease or not. It is a classification problem (at risk, not at risk). We use a classifier decision tree with a max depth of 4 levels, an entropy splitter function, balanced target class weights, and cost complexity pruning to classify a patient as either "at risk" or "not at risk". The model accuracy is 70%, a precision of 20.9%, a recall of 77.5%, and f1 of 33%. The model's confusion matrix and decision tree plot give more insight into how the model performs and along which features it splits the data. Future work would be to try a random forest model to try to improve the performance of predicting heart disease.

## Process Data

```python
from sklearn.model_selection import train_test_split # library to separate data into training data and testing data
```

```python
# feature variables, will become decision nodes within our decision tree
X = df.drop('HeartDiseaseorAttack', axis=1)
# target variable, what we are trying to predict
y = df['HeartDiseaseorAttack']
```

```python
# split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) 
```

## Configure Model

I played around a decent amount with the hyperparameters of the decision tree. These are the settings I found most generalizable and with interesting information. 

This is a classification problem (predict if a patient is in group "no heart risk" or "heart risk"), so we want to use the classifier decision tree rather than the regression decision tree. 

Our dataset is highly skewed, so I used entropy as the splitting function to select the feature to use on each node in the tree. Entropy is a measure of disorder or randomness in a system. We want to minimize entropy (ranges from 0 to 1) to try and separate out datapoints as best as possible at each node (mljourney, 2024). I chose entropy over the standard scikit-learn Gini Impurity because entropy handles imbalanced target classes (mljourney, 2024). 

The other way I tried to address a skewed dataset was with the class_weight hyperparameter. The relative weights (importance of misclassifying a datapoint) of the "no heart disease" class with the "heart disease class" is balanced by adjusting the weight to be inversely proportional to the class size (DecisionTreeClassifier, n.d.). 

To prevent overfitting the model to the training data (very common with decision trees), I limited the number of generations in the decision tree to 4. I also pruned off branches that do not add information to the model (for example, both leaves from a decision node classify into the same category) using cost complexity pruning (DecisionTreeClassifier, n.d.). The larger the pruning parameter, the more branches it cuts off to simplify the model. I chose 0.01 because it left an interesting tree that wasn't super big or complex. In the future, optimizing this hyperparameter could further improve the model. 

```python
from sklearn.tree import DecisionTreeClassifier # library for a decision tree machine learning algorithm
```

```python
# initiate decision tree model. This is a classification problem, so we use a classification decision tree rather than a regression decision tree
dtc = DecisionTreeClassifier(criterion='entropy', # decide how to split the dataset at each node by minimizing entropy, measure of disorder in system
                             max_depth=4, # limit number of generations in tree to 4, helps prevent overfitting
                             ccp_alpha=0.01, # cost complexity pruning: removes branches that don't add much information. Bigger numbers prunes more.
                             class_weight='balanced') # balance weights of each group (no heart problems vs heart disease) to address skewed dataset
```

## Train Model

```python
# fit decision tree with training data
dtc.fit(X_train, y_train)
```

## Test Model

```python
# predict heart disease in patients in the test set
y_pred = dtc.predict(X_test)
```

## Results

I tested out as many performance metrics as I had time for to start understanding how the metrics work and evaluate the performance of my model. 

```python
import numpy as np # library to do numerical calculations on arrays
```

```python
# Measure of accuracy. Ratio of correctly predicted results compared to the test test.
accuracy = np.sum(y_pred == y_test) / len(y_test)
print("Accuracy:", accuracy)
```

Accuracy measures the ratio of accurately predicted results. It is a very general picture, which does not show any skew in the data, such as always incorrectly predicting one specific category. Here, our accuracy is 69.9%. To get more details, we can use a confusion matrix that separates out each correctly predicted category and each incorrectly predicted category.

```python
# confusion matrix to see if there is scew in the accuracy test
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# define matrix, comparing real and predicted results
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat)

# visualize matrix with heatmap and descriptive labels
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=['No Heart Disease', 'Heart Disease'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

In the confusion matrix, the cells going left to right and top to bottom represent accurately predicted "No Heart Disease", falsely predicted "Heart Disease", falsely predicted "No Heart Disease", and accurately predicted "Heart Disease". We want to have higher rates (darker colors) along the left-to-right top-to-bottom diagonal. From the confusion matrix, it seems like we did pretty good with accurately predicting no heart disease, but we also miscategorized people as having heart disease when they did not. 

Let's look at the matrix in terms of ratios to better understand how the algorithm performed.

```python
# Look at percentage version of matrix
cm_normalized = confusion_matrix(y_test, y_pred, normalize='all')
print(cm_normalized)

# visualize matrix with labels
disp = ConfusionMatrixDisplay(confusion_matrix=cm_normalized, display_labels=['No Heart Disease', 'Heart Disease'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Normalized Confusion Matrix')
plt.show()
```

This shows that 62% of the data was accurately predicted as "No Heart Disease", 7.4% was accurately predicted as "Heart Disease", 28% was inaccurately classified as having heart problems, and 2.1% was inaccurately classified as not having heart problems. 

```python
# pandas version of confusion matrix
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
# precision and recall
from sklearn.metrics import precision_score, recall_score, f1_score # performance metrics library

# precision, ratio of correct positive predictions out of all the positive predictions
print("Precision:", precision_score(y_test, y_pred))

# recall, ratio of true positives accurately identified out of all the real positive values
print("Recall:",recall_score(y_test, y_pred))
```

Precision/Positive Predictive Value: "Proportion of positive cases correctly identified" (Srivastava, 2019). 
Recall/Sensitivity: Proportion of correctly identified positive cases (Srivastava, 2019).

Of all the patients our model predicted to have heart disease, 21% of them actually have heart disease. It tends to predict more people have heart disease than actually do. This is probably a good way to err, I'd rather people take risk of heart disease seriously and not be aware they could be at risk.

Of all the patients that have heart disease, our model correctly classified 77.5% of them.

```python
# F1 metric
f1_score(y_test, y_pred)
```

Combination of the precision and recall. We get an f1 score of 33%.

Our tree splits the dataset based on blood pressure on the root node. The entropy stays 1, so I feel like this doesn't actually make the data less random. I'm not sure why it is chosen. At every node, the entropy seems rather high. The lowest entropy seen is 0.47, which is still 47% mixed results.
