## Process Data

It is important to process the data for null values to ready it for machine learning. The above code line, from Gao Hongnan (2021) in Kaggle determines the number of null values in each column. We can see the columns age, cabin, embarked, boat, body and home.dest have missing values. Body refers to body identification number, which presumably is only for passengers who did not survive and whose body was found. It doesn't hold much predictive value for survival, so it is a column we can drop. Furthermore, passenger name may contain some information helpful for predicting survival, such as "Mr." and "Mrs.", but we can get that information from the sex column and pclass column. Extracting information from name will be challenging, so we will also drop this column. To process the null values, we will fill them with 0. There are more sofisticated ways to handle null values, such as filling them with the mean value, but this is a simpler way for the scope of this project.

```python
df.drop(labels=['body','name'], axis=1, inplace=True)
df.fillna(0, inplace=True)
print(df.head())
```

```python
df.info()
print('Columns with null values after fill NaN with 0:', df.isnull().sum(), sep = '\n')
```

We now have no null values and have dropped 2 columns from our data set. The next step is to convert columns with object data types to numbers so machine learning algorithms can more easily compare columns. Many of these columns have few unique values. For example, the sex column just contains values 'male' and 'female'. There are only 3 ports from which passengers embarked, C = Cherbourg; Q = Queenstown; S = Southampton. Therefore, we can encode these values into numerical representations. In this way, for the sex column, 'male' would be encoded 0 and 'female' would be encoded 1 (or vice versa). The function below encodes our categorical data. It is from Python Programming, K Means with Titanic Dataset. 

```python
# function definition to convert text data to numerical data that accepts dataframes as arguments
def handle_non_numerical_data(df):
    # create list of column names
    columns = df.columns.values
    # loop through each column in our column list
    for column in columns:
        #create an empty dictionary for inputting unique text values and map them to digit values
        text_digit_vals = {}
        # function to convert a value to an integer
        def convert_to_int(val):
            # return the digit mapped to the unique text from our dictionary
            return text_digit_vals[val]
       
        # check to see if the datatype of a column is not a number (neither integer nor float)
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            # if the above condition is met (not numbers):
            # convert the column values to a list
            column_contents = df[column].values.tolist()
            # create a set from the list of column contents. Set removes any duplicate elements, so we are left with only unique elements.
            unique_elements = set(column_contents)
            # initialize digit variable
            x = 0
            # iterate over every element in the unique element set
            for unique in unique_elements:
                # check if the element is in the dictionary defined above
                if unique not in text_digit_vals:
                    # if the above condition is met (not in the dictionary):
                    # set the dictionary value to the digit variable
                    text_digit_vals[unique] = x
                    # increase the digit variable by 1
                    x+=1

            # convert the list back to a dataframe column after mapping the convert_to_int function onto each value, which replaces non-number data with an integer
            df[column] = list(map(convert_to_int, df[column]))

    return df
```

We apply the above function to the dataframe and get the following results:

```python
df = handle_non_numerical_data(df)
print('after transforming dataset to all numeric')
print(df.head())
df.info()
```

We can see all the datatypes have been converted to integers or floats. Female has been coded to 1 and male has been coded to 0. Cabin numbers are simplified to pure numbers. Embarking ports and home/destinations have been coded to integers. All the above processing allows machine learning models to compare the data and find patterns, which would have been harder or impossible with null values and non-numeric data.

## Unsupervised Machine Learning with K Means

The first algorithm we will use is an unsupervised clustering model. This means that the data is not labeled or told how to group data. It simply looks at values and chooses clusters based on those values. With K Means clustering, the programmer gives the algorithms how many clusters they want as the only parameter. In our case we are predicting if a passenger survived or perished on the titanic, so we want 2 clusters: the survived group and the perished group.The computer then chooses two random data points as the centroids of these clusters, creates subsets in the data for the data points closest to each centroid, and determines the mean of each subset (Python Programming Tutorials, n.d.). The means become the new centroids. The algorithm recursively finds centroids until they stop moving, which represents the center of the data for each cluster. To predict which cluster a new data point will belong to, it determines which centroid that data point will be closest to and classifies it as belonging to that cluster. 

We need some libraries from skicit learn for K Means Clustering: the algorithm and preprocessing capabilities.

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

We need to specify the feature columns (everything except survival status) and the target column (survival). 

```python
# Process data for unstructured clustering
X = np.array(df.drop(labels = ['survived'], axis=1).astype(float))
y = np.array(df['survived'])
print(f'numpy array of X: {X}')
print(f'numpy array of y: {y}')
```

Next we want to create the model as a K Means clustering model with two clusters and fit the model with our feature variable data

```python
# unstructured clustering with K Means
clf = KMeans(n_clusters=2)
clf.fit(X)
```

The following code defines a function to measure the accuracy of the fitted model. The code is from Python Programming Tutorials, K Means with Titanic Dataset. The function compares the predicted group assignment of every data point with the actual group that data point belongs in. In the unsupervised model, the groups are randomly assigned as group 0 and group 1. In our data, the survived group is classified as 1 and the perished group is classified as 0. Therefore, the model may sometimes be the inverse of how the data is classified (70% accurate is the same as 30% accurate, because if you keep running the same code it will eventually switch the group names and be the opposite). 

```python
# Measure accuracy of model, accepts numpy array of feature variables as an input
def measure_accuracy(X):
    # initiate variable to tally how many correct predictions the model makes
    correct = 0
    # loop through every element in the numpy array
    for i in range(len(X)):
        # make every element a floating point data type
        predict_me = np.array(X[i].astype(float))
        #print(f'np array for {i} row: {predict_me}')
        # Reshape each row into a nested array. Makes it a 2D array that is for some reason necessary. Causes a value error if I comment it out 
        predict_me = predict_me.reshape(-1, len(predict_me))
        # print(f'numpy array for {i} row after reshaping: {predict_me}')
        # apply the fitted k means model to the row to predict which cluster that passenger belongs to
        prediction = clf.predict(predict_me)
        # if the predicted group for that row (passenger) is the same as the actual group for that passenger, which is the value of the survival column, which is stored in variable y
        if prediction[0] == y[i]:
            # add one to the correct tally
            correct += 1

    print('accuracy of predicting survival of passengers on titanic:')
    # find the ratio of accurate predictions by dividing the correct number of predictions by the total number of predictions
    print(correct/len(X))
```

```python
measure_accuracy(X)
```

The accuracy ratio is shown above. I got 0.51, or 51% accuracy in predicting which group a passenger belongs to. This isn't very high, basically just random. A way to improve the accuracy is to scale each feature so it has a mean of 0 and standard deviation of 1 (Scikit-Learn, n.d.). This keeps features with larger numbers from dominating the model. We use skicit learn's preprocessing library. 

```python
X = preprocessing.scale(X)
```

Refit the model with the scaled data and then measure it's accuracy:

```python
clf.fit(X)
```

```python
measure_accuracy(X)
```

We get an accuracy of 27%, which is the inverse of 73%. Because which group is assigned which number is random, we conclude that our k means clustering model with scaled data is 73% accurate at predicting which passengers died or survived on the titanic.

## Supervised Machine Learning with Linear Regression

The other main category of machine learning algorithms are supervised ML models. Supervised machine learning uses labeled data to train a model. The model is told what it is looking for (what the categories are, for example). The model is trained on one subset of the data and then tested with another subset of the data, whose target variable is known, to determine the accuracy of the model. 

The supervised model used for this project is linear regression. Regression models attempt to calculate a line of best fit for the data. We need some libraries from Skicit-Learn to do linear regression: a way to train the model, a way to determine the accuracy of the training, and the model itself. 

```python
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score 
```

First we want to set up the model.

```python
# supervised learning
model = LinearRegression()
```

Now we need to train the model on our dataset. We could train it one one subset and test it on the rest, but that can lead to overfitting the data to the training set. Also, variability in the training/testing data could make a model look like it is performing really well, but then perform worse on new data. A way to get a better sense of how accurate your model is, is to train it multiple times on multiple sets of your data. Say you divide your data into 4 subsets. You can use 3 subsets to train the data and 1 subset to test it, then rotate which subsets are the training and testing sets 4 times. Each iteration will give you the accuracy for that arrangement. The average accuracy, and the variation between each arrangement, gives you an overall idea of how well this type of model will predict this type of data. 

Cross_val_predict uses this rotating arrangement of training and testing data to evaluate models. We will give it the model (linear regression), the feature data (X) and the data to predict (y).

We can then calculate two important metrics on model accuracy: mean squared error and the coefficient of determination (r2). Mean squared error is the average distance every data point lies from the line of best fit (Arcitura Education, 2021). The mean squared error here is the average across every model of those averages for each model. We want this number to be close to 0, which represents data points lying very close to the line of best fit. The coefficient of determination is a measure of how correlated the target variable is to the predictive variables (Arcitura Education, 2021). In this scenario, it is the percent of the variation in 'survival' that can be explained by variation in the feature variables. The closer to 1 (100%), the more variation in survival is explained by the feature variables. 

```python
# run cross validation to get estimate on accuracy and variance of predictions
estimates = cross_val_predict(model, X, y)
print(estimates)

# calculate metrics, how accurate and variable model is
mse = mean_squared_error(y, estimates)
r2 = r2_score(y, estimates)

print("Cross-Validation MSE:", mse)
print("Cross-Validation R:", r2)
```

The MSE is 0.08 and the coefficient of determination is 66%. This means that the average distance data points lie from the line of best fit is 0.08 (seems pretty close), and that 66% of the change in survival is explained by change in the feature variables.
